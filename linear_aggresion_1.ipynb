```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("D:/junotbok/5th week/SLR_house_prices.csv")
df.head(20)
```

```python
df.shape
```

```python
df.info()
```

```python
area_mean = df['area'].mean()
```

```python
price_mean = df['price'].mean()
```

```python
df['area']= np.where(df['area'].isnull(),area_mean,df['area'])
```

```python
df['price']= np.where(df['price'].isnull(),area_mean,df['price'])
```

```python
df.info()
```

```python
x,y = df[['area']], df[['price']]
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.3)
```

```python
print(train_x.shape, train_y.shape, test_x.shape, test_y.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(train_x,train_y)
```

```python
#pip install scikit-learn
```

```python
pred=model.predict(test_x)
```

```python
plt.plot(range(0,len(test_y)),test_y)
plt.plot(range(0,len(test_y)),pred)
plt.show()
```

```python
plt.plot(range(0,10),test_y[0:10])
plt.plot(range(0,10),pred[0:10])
plt.show()
```

```python
mean_absolute_error(test_y,pred)
```
