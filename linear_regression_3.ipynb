```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("D:/junotbok/5th week/experience_salary_data_updated.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
x,y = df[['Years of Experience']] , df[['Salary']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.3)
```

```python
print(train_x.shape, train_y.shape, test_x.shape, test_y.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(train_x, train_y)
```

```python
pred=model.predict(test_x)
```

```python
plt.plot(range(0,len(test_y)),test_y)
plt.plot(range(0,len(test_y)),pred)
plt.show()
```

```python
plt.plot(range(0,10),test_y[0:10])
plt.plot(range(0,10),pred[0:10])
plt.show()
```
